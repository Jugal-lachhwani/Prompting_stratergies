## The DSPy Philosophy: Programming vs. Prompting
In traditional prompting, changing a model (e.g., from GPT-5.5 to Claude Sonnet 4.6 or Llama 4) requires re-writing all your prompts. DSPy separates Logic from Formatting.

## 5. DSPy Module

This is where real applications are built.

```python
class MySystem(dspy.Module):

    def __init__(self):

        self.search = dspy.ChainOfThought(
            SearchQuery
        )

        self.answer = dspy.ChainOfThought(
            AnswerQuestion
        )

    def forward(self, question):

        query = self.search(
            question=question
        ).query

        docs = retrieve(query)

        return self.answer(
            context=docs,
            question=question
        )
```

## 7. Metric

```python
def exact_match(
    example,
    prediction,
    trace=None
):
    return (
        prediction.answer ==
        example.answer
    )
```
